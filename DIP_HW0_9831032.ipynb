```python
#@title Q2 - Part 1 (100%)
def get_matrix(dims: tuple[int, int], seed: int) -> np.ndarray:
    # check validity of 'dims' input
    assert (
        isinstance(dims, tuple)
        and len(dims) == 2
        and isinstance(dims[0], int)
        and isinstance(dims[1], int)
    ), "'dims' is not a tuple of length two with int type"

    # check validity of 'seed' input
    assert (
        isinstance(seed, int)
    ), "'seed' is not an int type"

    # initialize the matrix
    _matrix = np.zeros(dims, dtype=int)

    # set default parts in all matrices
    _matrix[0, :], _matrix[:, 0] = seed, seed
    _matrix[:, 0][1::2] = -seed

    # set other parts
    for i in range(1, dims[0]):
        for j in range(1, dims[1]):
            _matrix[i][j] = _matrix[i][j-1] - _matrix[i-1][j] - _matrix[i-1][j-1]

    return _matrix

# examples
print(get_matrix((3, 4), 1), '\n')
print(get_matrix((5, 3), 3))
```

```python
#@title Q3 - Part 1 (0%)
std_num = '9831032'
```

```python
#@title Q3 - Part 2 (40%)
def get_circle(r: int) -> np.ndarray:
    # array size
    d = (2 * r) - 1
    # x and y range in the array
    y, x = np.ogrid[:d, :d]
    # center of our circle in the array
    center = (r - 1, r - 1)
    # initialize circle
    circle = np.zeros((d, d), dtype=np.uint8)
    # find indices of the circle inside the array (using equation: x^2 + y^2 < r^2)
    mask = (x - center[0]) ** 2 + (y - center[1]) ** 2 <= (r - 1) ** 2
    # set value in desired indices of array to show the circle
    circle[mask] = 255

    return circle

# examples
circle_3, circle_4 = get_circle(3), get_circle(4)
print(circle_3, '\n\n', circle_4)
```

```python
#@title Q3 - Part 3 (35%)
def noise_creation(_array, domain):
    # create noise array in the same shape of origin array
    noise = np.random.uniform(0, domain, _array.shape)

    # change noise for elements of the array that are equal to 255
    non_zero_indices = np.where(_array == 255)
    noise[non_zero_indices] *= -1

    # add noise and round
    _array = _array + noise
    _array = np.floor(_array)
    _array = _array.astype(np.uint8)

    return _array

noise = 20 + sum([int(d) for d in std_num if d.isdigit()]) % 15

# examples
noisy_circle_3 = noise_creation(circle_3, noise)
noisy_circle_4 = noise_creation(circle_4, noise)
print(noisy_circle_3, '\n\n', noisy_circle_4)
```

```python
#@title Q3 - Part 4 (15%)

def img_show(original, noisy):
    # show both plots in a window
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8, 6))

    # set figure title
    fig.suptitle(f'HW0_Image_{std_num}', fontsize=16)

    # plot data on the first subplot
    ax1.imshow(original, cmap='gray')
    ax1.set_title('Original circle')
    ax1.axis('off')

    # plot data on the second subplot
    ax2.imshow(noisy, cmap='gray')
    ax2.set_title(f'Noisy circle (domain: {noise})')
    ax2.axis('off')

    # adjust spacing between subplots
    plt.tight_layout()

    # show the plots
    plt.show()

# example
img_show(circle_4, noisy_circle_4)
```

```python
#@title Q3 - Part 5 (10%)

def img_show_3d(original, noisy):
    # create a grid
    x = np.linspace(0, original.shape[0], original.shape[0])
    y = np.linspace(0, original.shape[1], original.shape[1])
    X, Y = np.meshgrid(x, y)

    # create a 3D plot
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5), subplot_kw={'projection': '3d'})

    # set figure title
    fig.suptitle(f'HW0_Surface_{std_num}', fontsize=16)

    # plot the 3D surface for original image
    ax1.plot_surface(X, Y, original, cmap='gray')
    ax1.set_title('Original circle')

    # plot the 3D surface for original image
    ax2.plot_surface(X, Y, noisy, cmap='gray')
    ax2.set_title(f'Noisy circle (domain: {noise})')

    plt.show()

# example
img_show_3d(circle_4, noisy_circle_4)
```

```python
#@title Q4 - Part 1 (15%)

# Load the PNG image using OpenCV
image_path = 'chest-xray.png'
image = cv2.imread(image_path)

# OpenCV reads images in BGR format, we need to convert it to RGB
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
print(f'RGB format of the image: {image_rgb.shape}')

# convert to grayscale
image_gray = cv2.cvtColor(image_rgb, cv2.COLOR_RGB2GRAY)
print(f'Grayscale format of the image: {image_gray.shape}\n')

# display the original image and grayscale image side by side
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(image_rgb)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(image_gray, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')

plt.show()
```
